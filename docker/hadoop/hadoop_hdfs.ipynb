```python
import socket

# 현재 서버의 호스트네임과 IP 확인
hostname = socket.gethostname()
ip_address = socket.gethostbyname(hostname)

print(f"현재 호스트네임: {hostname}")
print(f"현재 내부 IP 주소: {ip_address}")
```
